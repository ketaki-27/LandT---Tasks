```python
# converting dictionary into a dataframe
df = pd.DataFrame(data=dataset.data,columns=dataset.feature_names)
df.insert(64,'target',dataset.target)
```

```python
# storing the values into x and y variables
X = df.iloc[:,:-1]
```

```python
y = df["target"]
y
```

# Before PCA

```python
# scaling the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
X_scaled
```

```python
X.shape,y.shape
```

```python
# splitting the data into training and testing dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=30)
```

```python
# training the model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

# After PCA

```python
from sklearn.decomposition import PCA
# 95 % accuracy
pca =PCA(0.95)
X_pca =pca.fit_transform(X) 
```

```python
X.shape
```

```python
X_pca.shape  
```

```python
# calculating the covariance matrix
print("Covariance matrix :\n",pca.get_covariance())
```

```python
# calcualting the eigen vectors and eigen values
eigen_vectors = pca.components_
eigen_values = pca.explained_variance_
```

```python
pca.n_components_
```

```python
print("Eigen vectors :\n",eigen_vectors)
```

```python
print("Eigen vectors :\n",eigen_values)
```

```python
pca.explained_variance_ratio_
```

```python
#  Visualize explained variance ratio
import numpy as np
explained_variance_ratio = pca.explained_variance_ratio_
plt.figure(figsize=(8, 6))
plt.plot(np.cumsum(explained_variance_ratio))
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('Explained Variance Ratio for Different Numbers of Principal Components')
plt.grid(True)
plt.show()
```

```python
#  Project data onto the first two principal components
# Visualize reduced representation
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('Projection of Data onto First Two Principal Components')
plt.colorbar(label='Digit Label')
plt.grid(True)
plt.show()
```

```python
# splitting the data again after pca
X_train_pca,X_test_pca ,y_train,y_test = train_test_split(X_pca,y,test_size=0.2,random_state=0)
```

```python
# training the data again using pca 
model = LogisticRegression()
model.fit(X_train_pca,y_train)
model.score(X_test_pca,y_test)
```
